```python
import pandas as pd
data = pd.read_csv("C:/Users/j/Desktop/Dataset/GenZ_DatingApp_Data.csv")
df =data.drop(columns=["User_ID","Primary_App","Secondary_Apps","Usage_Frequency","Daily_Usage_Time","Challenges","Desired_Features","Preferred_Communication","Partner_Priorities"])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x=df_fresh["Gender"],y=df_fresh["Age"])
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("First Plot")
plt.show()
```

```python
df_agg = df_fresh.groupby("Gender")["Age"].mean().reset_index()
plt.figure(figsize=(10,6))
plt.bar(df_agg["Gender"],df_agg["Age"])
plt.xlabel("GENDER")
plt.ylabel("AGE")
plt.title("THE PLOT")
plt.show()
```

```python
df_fresh2 = df_fresh.groupby("Satisfaction")["Age"].mean().reset_index()
plt.figure(figsize=(10,6))
plt.scatter(df_fresh2["Satisfaction"],df_fresh2["Age"],color="Red")
plt.xlabel("AGE")
plt.ylabel("SATISFACTION")
plt.title("HAHA")
plt.show()
```

```python
df3= df_fresh.groupby("Reason_for_Using")["Age"].mean().reset_index()
plt.figure(figsize=(10,6))
plt.bar(df3["Reason_for_Using"],df3["Age"])
plt.xlabel("REASON")
plt.ylabel("AGE")
plt.title("Variation")
plt.show()
```
