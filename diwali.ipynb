# Exploratory Data Analysis

### Gender

```python
import matplotlib.pyplot as plt
import seaborn as sns
# plt.figure(figsize=(10,6))
ax = sns.countplot(x="Gender",data=data,palette="rainbow")
for bars in ax.containers:
    ax.bar_label(bars)
```

### Plotting a bar chart for Gender vs Total amount

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,palette="spring")
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age Group 

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Age Group",hue="Gender",data=data,palette="rainbow")
for i in ax.containers:
    ax.bar_label(i)
    
```

*It seems like people of age group 26 to 35 are buying the most products,specificly female are buying more than male in any group of age(According to count).*

### Total amount vs Age Group

```python
sales_age = data.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
x=sns.barplot(x="Age Group",y="Amount",data=sales_age,palette=["#FF6B6B", "#4ECDC4", "#45B7D1","#FAD02E", "#594F4F"])
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female(According to amount).*

### State vs Orders

```python
#Total number of orders from top 10 States
sales_state = data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,6)})
barplot=sns.barplot(x="State",y="Orders",data=sales_state,palette="viridis")
# plt.grid(visible=False)

for label in barplot.patches:
    barplot.annotate(format(label.get_height(),"0.1f"),
                     (label.get_x()+label.get_width()/2,label.get_height()),
                     ha="center",va="center",
                     xytext=(0,5),
                     textcoords="offset points")
```

*Most orders from Uttar Pradesh,Maharashtra and Karnataka.*

### Total amount vs sales from top 10 states.

```python
sales_amount =data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

amount=sns.barplot(x="State",y="Amount",data=sales_amount,palette="bright")
# plt.grid(visible=False)

for i in amount.patches:
    amount.annotate(format(i.get_height(),"0.1f"),(i.get_x()+i.get_width()/2,i.get_height()),
                    ha="center",va="center",
                    xytext=(0,5),
                    textcoords="offset points")
```

### Pie chart

```python
sales_amount =data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(4)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0.3, 0)
size=sales_amount["Amount"]
labels=sales_amount["State"]
plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

### Marital Status

```python
marital_status= data.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(6,5)})
sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=marital_status,palette="magma")
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={"figure.figsize":(20,6)})
ax=sns.countplot(x="Occupation",data=data,palette="spring")
plt.xlabel("Count")
plt.ylabel("Occupation")
plt.title("Occupation vs Count")
for j in ax.containers:
    ax.bar_label(j)
```

*Most of the buyers works on IT sector,Healthcare and Banking Count.*

### Occupation vs Amount

```python
occu_amo =data.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(x="Occupation",y="Amount",data=occu_amo,palette="Accent")
for count in ax.patches:
    ax.annotate(format(count.get_height(),"0.1f"),
                (count.get_x()+count.get_width()/2,count.get_height()),
                ha="center",va="center",
                xytext=(0,5),
                textcoords="offset points")
```

*From above graphs we can see IT, Healthcare and Aviation sectors people spend more money than other sectors.On the another hand Agriculture,Textile,Constuction sectors people spend less money compare to other sectors.*

### Product Category

```python
sns.countplot(x="Product_Category",data=data,palette="plasma")
```

*It seems like most of the product is Clothing & Apparel,Food and Electronics & Gadgets.*

### Product Category vs Amount

```python
pc_amount= data.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(5)
plt.pie(pc_amount["Amount"],explode=(0,0,0.5,0,0.3),labels=pc_amount["Product_Category"],
        colors=["red","green","blue","yellow","orange"],shadow=True,autopct="%1.1f%%",startangle=140)
plt.axis("equal")
plt.title("Pie Chart of top 5 Products Amount")
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

### Conclusion

*Married women age group 26-35 yrs from Uttarpradesh,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.*
