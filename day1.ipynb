## What is the most commonly awarded gender and birth country?

```python
top_gender = nobel['sex'].value_counts().index[0]
top_country = nobel['birth_country'].value_counts().index[0]
print(f'The most commonly awarded gender is {top_gender} and country is {top_country}')
```

## Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

```python
nobel['decade'] = round(nobel['year']/10)*10
nobel['america'] = nobel['birth_country']==  'United States of America'
highest_ratio = nobel.groupby('decade',as_index=False)['america'].mean()#as_index =False makes the data like a table
max_decade_usa = highest_ratio[highest_ratio['america']==highest_ratio['america'].max()]['decade'].values[0]#keep eyes on here for the logic
print(f'{max_decade_usa} had the highest ratio of US-born Nobel Prize winners to total winners in all categories')
```
