# Job Postings by Sector per Month

```python
sector_map = {
    4: 'mf',  # manufacturing
    6: 'con',  # construction
    7: 'co', # commerce
    8: 'sv',  # services
   
}
data['econ_sector'] = data['sector'].map(sector_map)

monthly_total = data.groupby('date').size().reset_index(name='total_postings')
monthly_by_sector = (
    data.groupby(['date', 'econ_sector'])
      .size()
      .reset_index(name='postings')
)



monthly_sector_pivot = monthly_by_sector.pivot(
    index='date',
    columns='econ_sector',
    values='postings'
).fillna(0)
monthly_sector_pivot
```

```python
import matplotlib.pyplot as plt

# Total job postings over time
plt.figure(figsize=(10, 4))
plt.plot(monthly_total['date'], monthly_total['total_postings'], marker='o')
plt.title("Total Job Postings Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Postings")
plt.grid(True)
plt.tight_layout()
plt.show()

# Job postings by sector
monthly_sector_pivot.plot(figsize=(12, 6), marker='o')
plt.title("Monthly Job Postings by Sector")
plt.xlabel("Date")
plt.ylabel("Number of Postings")
plt.legend(title="Sector")
plt.grid(True)
plt.tight_layout()
plt.show()

```

#  Seasonal Patterns

```python
# Extract month and group across all years
data['month_name'] = data['date'].dt.month_name()
monthly_seasonality = data.groupby('month_name').size().sort_index()

# Normalize if needed (e.g., divide by number of years)
monthly_avg = monthly_seasonality / data['date'].dt.year.nunique()
monthly_avg.plot(kind='bar', title="Average Job Postings by Month");
```

# Growth / Decline Trends

```python
yearly_postings = data.groupby('date').size()

# Compute year-over-year growth
yearly_growth = yearly_postings.pct_change() * 100
print(yearly_growth)
yearly_growth.plot(title='Yearly Growth')
```

#  Sector-Specific Trends

```python
sector_yearly = data.groupby(['date', 'econ_sector']).size().unstack().fillna(0)
sector_growth = sector_yearly.pct_change() * 100
print(sector_growth)
sector_growth.plot(title='Sector Growth')
```

#  Sectoral Volatility

```python
# Create monthly sector trend
monthly_sector = data.groupby(['date', 'econ_sector']).size().unstack().fillna(0)

# Calculate standard deviation as proxy for volatility
sector_volatility = monthly_sector.std().sort_values(ascending=False)
print(sector_volatility)
sector_volatility.plot(title='Sector Volatility')
```

* High std dev = sector is highly sensitive or seasonal

# Dominant sectors by platform and month:

```python

#Aggregate job postings by source, sector, and month
platform_sector_month = data.groupby(['empresa', 'sector', 'month_year']).agg({'job_postings': 'sum'}).reset_index()

#Identify dominant sectors for each platform and month
dominant_sectors = platform_sector_month.loc[
    platform_sector_month.groupby(['empresa', 'month_year'])['job_postings'].idxmax()
]

print("Dominant sectors by platform and month:")
# print(dominant_sectors)

#Compare sectoral distribution across platforms
sector_distribution = data.groupby(['empresa', 'sector']).agg({'job_postings': 'sum'}).reset_index()

sector_distribution
```
